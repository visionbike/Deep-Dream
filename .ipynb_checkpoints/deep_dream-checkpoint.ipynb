## Import

```python
from matplotlib import pyplot as plt
%matplotlib inline
from tensorflow.python.keras.layers import Input
from preprocess import *
from model import *
from scipy.optimize import fmin_l_bfgs_b
```

Set up early settings.

```python
# The model will not be trained, this command is used to disable all training-specific operations
K.set_learning_phase(0)

# Image settingd
im_path = 'base_image.jpeg'
im_width = 1200
im_height = 900
im_size = (im_height, im_width, 3)

# Tensor to store the model's input
dream = Input(batch_shape=(1,) + im_size)

model_name = 'inception_v3'
```

## 1. Deep Dream with InceptionV3 network

The original Deep Dream release was an Inception V3 model. We use InceptionV3 model that comes with Keras.

### Build InceptionV3 network

```python
# Build the InceptionV3 network
# The model will be loaded with pre-trained ImageNet weights.
model = get_inceptionv3_model('imagenet', input_tensor=dream)
```

List all layers names using `model.summary()`.

```python
model.summary()
```

### Define loss and gradients

We will tell the contributions of the layers in the loss.

```python
# Dict mapping layer names to a coefficient quantifying how much the layer's activation
# will contribute to the loss we will seek to maximize.
# Note that these are layer names as they appear in the built-in InceptionV3 application.
layer_contributions = {'mixed2': 0.05,
                       'mixed3': 2.,
                       'mixed4': 0.2,
                       'mixed5': 1.5}

# The continuity loss weight, to give the image local coherence and avoid messy blurs
continuity = 0.5
# The L2 norm loss weight in order to prevent pixels from taking very high values
dream_l2 = 0.5
jitter = 0.05

# List of layer in the model
layer_dict = dict([(layer.name, layer) for layer in model.layers])
```

Now, we define a tensor the contains loss.

```python
loss_and_grads = define_loss_and_grads(layer_dict, dream, im_size,layer_contributions, continuity, dream_l2)
```

### Run DeepDream

Load image.

```python
# Load the image into a Numpy array for inception_v3 network
im = preprocess_image(im_path, im_size, model_name)
```

Now is to run L-BFGS optimizer over the pixels of the generated image, in order to minimize the loss.

```python
evaluator = Evaluator(loss_and_grads)
iterations = 30
interval = 6
im_save = im

for i in range(iterations):
    random_jitter = (jitter*2)*(np.random.random(im_size))
    im += random_jitter
    im, min_val, info = fmin_l_bfgs_b(evaluator.loss, im.flatten(),
                                     fprime=evaluator.grads, maxfun=7)
    print('Loss at {}: {}'.format(i, min_val))
    im = im.reshape(im_size)
    im -= random_jitter
    im_save = deprocess_image(np.copy(im), im_size, model_name)
    if (i + 1) % interval == 0:
        plt.imsave('dream_{}_at_iter_{}.png'.format(model_name, i), im_save)
        plt.imshow(im_save)
        plt.show()
plt.imsave('dream_{}_final.png'.format(model_name), im_save)
plt.imshow(im_save)
plt.show()
```
